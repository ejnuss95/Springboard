The international players in the league is about 10%. This is no suprise here given it is American league but I assume we will see that is has grown over time since 1992. 

```python
#I want to how the game has grown for American and International Players since the beginning of the dataset
american_number = basketball_cleaned_data[basketball_cleaned_data['US or International']=='US']['Season'].value_counts().sort_index()
international_number = basketball_cleaned_data[basketball_cleaned_data['US or International']=='International']['Season'].value_counts().sort_index()
```

```python
# Plot 
plt.plot(american_number, color = 'g', label = 'American Players')
plt.plot(international_number, color = 'm', label = 'International Players')
plt.plot(american_number + international_number, color = 'b', label = 'Total Amount of Players')
plt.legend(loc='upper left')
plt.title('Number of Players Total in the NBA')
plt.xlabel('Seasons')
plt.ylabel('Players playing in that year')
plt.show()
```

We can see American players is parallel to Total but we can see a steady increase in International Players as well. The increase is so large due to the amount of expanded teams and rosters over time.

##  Years of Experience in the league

```python
plt.rcParams['figure.figsize'] = [18, 8]
sns.swarmplot(x='Experience', y = 'PTS', data = basketball_cleaned_data, hue = 'US or International', palette = {'US': 'r', 'International': 'g'})
plt.title('PPG by Experience')
plt.show()
```

```python
plt.rcParams['figure.figsize'] = [18, 8]
sns.swarmplot(x='Draft Selection Pick', y = 'PTS', data = basketball_cleaned_data, hue = 'US or International', palette = {'US': 'r', 'International': 'g'})
plt.title('PPG by when you were drafted')
plt.show()
```

I cannot get really much out of these 2 swarmplots to see any sort of conclusion between US and international players. It looks like the higher you were drafted, the more points per game you are likely to have. It seems that players peak in PPG from years 5-8 in the league.

This is the overall average minutes per game. It is about just half of the game. I am going to do a box plot to see if we can find outliers among points per 21 minutes by age. 

```python
basketball_cleaned_data['PTS per 21m'] = basketball_cleaned_data['PTS'].divide(basketball_cleaned_data['MP'])*21
sns.boxplot(x='Age', y = 'PTS per 21m', data = basketball_cleaned_data, color = 'b')
plt.title('Points per 21 minutes by Age')
plt.show()
```

There are some outliers in here from ages 22 - 25 that scored a lot more points in those 22-25 minutes but this does not show us much in terms of what we are trying to find.

```python
plt.rcParams['figure.figsize'] = [18,10]
sns.heatmap(data=basketball_cleaned_data.corr())
plt.title('Heatmap')
plt.show()
```

```python
sns.lmplot('Draft Selection Pick', 'PTS', data = basketball_cleaned_data, size = 10, aspect=1.5, scatter_kws={'color':'black'},)
plt.plot([8/x for x in np.arange(.2,20,.1)], linewidth=5, color = 'blue')
plt.xticks([x for x in range (0, 30, 10)])
plt.xlim(-1, 30)
plt.ylim(-1, 30)
plt.show()
```

This lmplot is a better representation than the swarm plot but this reiterates that there is an inverse relationship between draft selection and points per game. 

```python
sns.lmplot(x='GS', y = 'PTS', data = basketball_cleaned_data, size = 10, scatter_kws = {'color':'black'}, line_kws={'linewidth': 5})
plt.title('Games started vs points per game')
plt.show()
```

We see a positive relationship amongst games started and points per game. This makes sense given the more you start, the more you play, and the more likely you are to score. 

```python
sns.lmplot(x='FG%', y='PTS', data = basketball_cleaned_data, size = 10, aspect = 1.5, scatter_kws = {'color': 'black'}, fit_reg = False)
```

This is FG% versus points per game. We can see on average, people make about 50% of their shots when scoring a lot of points in a game. We have some outliers in the beginning and front who score 0 and 100% in a game but its usually very little points. 

```python
sns.lmplot(x='FT%', y = 'PTS', data = basketball_cleaned_data, size = 10, scatter_kws = {'color':'black'}, line_kws={'linewidth': 5})
plt.title('FT% vs points per game')
plt.show()
```

We see somewhat positive relationship with Free Throw percentage and PPG. This is not perfect positive. For example, big men like Shaquille O'Neal and Dwight Howard are great scorers and defenders but not great free throw shooters.
