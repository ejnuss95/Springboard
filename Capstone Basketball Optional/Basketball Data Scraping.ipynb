# Data Scraping

```python
#Here I am webscraping data from the basketball reference website in order to create our dataset. 
import re
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
from urllib.request import urlopen 


#These are my two urls that I am scraping from
original_url ="https://www.basketball-reference.com"
player_url="https://www.basketball-reference.com/players/"
soups={}

#I must go through every letter in the alphabet except X, there are no players who have had the last name X
for i in range(26):
    if(i!=23): # 'x' webpage does not exist
        soups[chr(97+i)]=BeautifulSoup(urlopen(player_url + chr(97+i) + "/"))
       
    # Here I am creating an emply list where the players will go
    player_urls = []


# Here I am adding the player URLS to players list 
for key, letter_soup in soups.items():
    # This is just to indicate the first row is a header with all the different categories
    player_rows=letter_soup.table.find_all("tr")[1:]
    counter=0
    
    #I indicated players after 1992, the game became more global when Jordan and the Bulls became successful
    for idx, player in enumerate(player_rows):
        if int(player.td.string) >= 1992:
            player_urls.append(original_url + player.th.a["href"])
            counter+=1
                   
        
player_data_df= player_df=pd.read_html(str(BeautifulSoup(urlopen(player_urls[0])).table))[0].iloc[:-1, 0:30]
dataframe_index=1

countries=np.array(["albania","andorra","armenia","austria","azerbaijan","belarus","belgium",
                    " bosnia and herzegovina","bulgaria","croatia","cyprus","czech republic",
                    "denmark","estonia","finland","france","georgia","germany","greece","hungary",
                    "iceland","ireland","italy","kazakhstan","kosovo","latvia","liechtenstein",
                    "lithuania","luxembourg","macedonia","malta","moldova","monaco","montenegro",
                    "netherlands","norway","poland","portugal","romania","russia","san marino",
                    "serbia","slovakia","slovenia","spain","sweden","switzerland","turkey"])
states=np.array(["alaska", "alabama", "arkansas", "american samoa", "arizona", 
                 "california", "colorado", "connecticut", "district of columbia", "delaware", "florida", 
                 "georgia", "guam", "hawaii", "iowa", "idaho", "illinois", "indiana", "kansas", "kentucky", "louisiana", 
                 "massachusetts", "maryland", "maine", "michigan", "minnesota", "missouri", "mississippi", "montana", 
                 "north carolina", "north dakota", "nebraska", "new hampshire", "new jersey", "new mexico", "nevada", 
                 "new york", "ohio", "oklahoma", "oregon", "pennsylvania", "puerto rico", "rhode island", "south carolina", 
                 "south dakota", "tennessee", "texas", "utah", "virginia", "u.s. virgin islands", "vermont", "washington", 
                 "wisconsin", "west virginia", "wyoming"])



for player_url in player_urls[1:]: 
    try: # Just checking for any mistakes
        player_soup=BeautifulSoup(urlopen(player_url))
    except:
        print("Exception raised for player %d" % (dataframe_index))
        player_soup=BeautifulSoup(urlopen(player_urls))
    
    player_df=pd.read_html(str(player_soup.table))[0].iloc[:-1, 0:35]
    
    # I thought it was relevant to add how many years they have been in the league 
    player_df["Experience"]=player_df.index.values+1
   
    
    # Adding the draft placing of the player
    draft_link=player_soup.find("a", {"href":re.compile("/teams/.../draft.html")})
    if draft_link!=None:
        draft_string=list(draft_link.parent.stripped_strings)[2]
        player_df["Draft Selection Pick"]=draft_string[draft_string.find("pick, ")+6:draft_string.find(" overall")-2]
    else:
        player_df["Draft Selection Pick"]=-1
        
    country_or_state=list(player_soup.find("span", {"itemprop":"birthPlace"}).stripped_strings)[1].lower()
    player_df["Country or State"]=country_or_state
    US_or_EU="US"
    if country_or_state in countries:
        US_or_EU="International"
    if country_or_state in states:
        US_or_EU="US"
    player_df["US or International"]=US_or_EU

    player_data_df=player_data_df.append(player_df)
    print("%d/%d dataframes appended" % (dataframe_index, len(player_urls)-1))
    dataframe_index+=1

# Dropping the league column as they are all NBA players
player_data_df.drop(["Lg"], axis="columns", inplace=True)

# Writing the 'player data' dataframe to a .csv file.
player_data_df.to_csv("Basketball data.csv")
```
