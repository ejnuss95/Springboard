```python
#Cleaning Up the data
import numpy as np
import pandas as pd
with open("C:/Users/Bsnus/Desktop/Springboard-Projects/Capstone Project 2 - NBA/Basketball Data.csv", "r") as basketball_data:
    basketball_data_df=pd.read_csv(basketball_data)
```

```python
#Dropping an unnecessary column
basketball_data_df.drop(["Unnamed: 0"], axis="columns", inplace=True)
```

```python
for column in ["Tm", "Country or State", "US or International", "Experience", "G", "GS"]:
    basketball_data_df[column]=basketball_data_df[column].astype("category") 
```

```python
for column in ["Tm", "Country or State", "US or International", "Experience", "G", "GS"]:
    basketball_data_df[column]=basketball_data_df[column].astype("category") 
```

```python
# Labelling missing values
basketball_data_df=basketball_data_df.replace(-1, np.nan)
```

```python
# Needed to fix some countries in the 'US or International '
basketball_data_df.loc[basketball_data_df["Country or State"]=="fyr macedonia","US or International"]="International"
basketball_data_df.loc[basketball_data_df["Country or State"]=="bosnia and herzegovina","US or International"]="International"
```

```python
# Had to rid of some null values in the US or International
basketball_data_df=basketball_data_df.iloc[9:,:].reset_index(drop=True)
# Had to clean up the 2P columns as well
basketball_data_df=basketball_data_df[basketball_data_df["2P"].notnull()].reset_index(drop=True)
```

```python
#Keepign a copy of the df 
basketball_data_df_og=basketball_data_df.copy(deep=True)
```

```python
#More unnecessary columns with null values
basketball_data_df.drop(["3P%", "FT%"], axis=1, inplace=True)
basketball_data_df=basketball_data_df[basketball_data_df["2P%"].notnull()].reset_index(drop=True)
```

```python
# Converting ratios to percentages
basketball_data_df["2P%"]=basketball_data_df["2P%"]*100
basketball_data_df["eFG%"]=basketball_data_df["eFG%"]*100
```

```python
basketball_data_df = basketball_data_df.dropna()
```
