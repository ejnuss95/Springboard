## Formula to calculate distance of closest police station

```python
%time
from timeit import Timer
import itertools
from math import radians, sin, cos, asin, sqrt, pi, atan2

distance = []
earth_radius_miles = 3958.8

x = chicago['Location']
y = station_array

def shortest(needle, haystack):
    dlat = np.radians(haystack[:, 0]) - radians(needle[0])
    dlong = np.radians(haystack[:, 1]) - radians(needle[1])
    a = np.square(np.sin(dlat/2)) + cos(radians(needle[0])) * np.cos(np.radians(haystack[:,0])) * np.square(np.sin(dlong/2))
    
    circle_distance = 2 * np.arcsin(np.minimum(np.sqrt(a), np.repeat(1, len(a))))
    d = earth_radius_miles * circle_distance
    return np.min(d)
```

```python
def donumpy():
    shortest(x, y)
```

```python
for i in x:
    distance.append(shortest(i, y))
```
