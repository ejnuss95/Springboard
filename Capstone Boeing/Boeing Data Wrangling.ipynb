```python
# Option Volatilities
options = pd.read_csv('VOL-BA.csv')
```

This option data will provide historical volatility of stock and implied volatility of options during different maturity lengths.

```python
options.head()
```

```python
options['Date'] = pd.to_datetime(options.Date)
```

```python
options.shape
```

```python
options.tail()
```

In quandl, the stock data only goes back to 2013 and the Option data goes back to 2002. We will only use information where they overlap. 

```python
options = options.drop(columns = ['Hv60', 'Hv90', 'Hv120', 'Hv150', 'Hv180', 'Phv60', 'Phv90', 'Phv120', 'Phv150', 'Phv180'])
options.tail()
```

```python
#new df with only columns without NA
options_clean = options.dropna(axis = 'columns')
options_clean.head()
```

```python
options.info()
```

```python
na_options = options.isna()
na_options.head()
```

```python
options[pd.isnull(options).any(axis=1)]
```

```python
clean_options = pd.DataFrame(options)
clean_options = clean_options.set_index('Date')
clean_options.head()
```

```python
clean_options.tail()
```

```python
clean_options1 = clean_options['December-31-2017': 'Jan-01-2013']
clean_options1.head()
```

```python
prices = pd.DataFrame(BA)
prices = prices[['Close', 'Date']]
prices = prices.set_index('Date')
prices.head()
```

```python
prices1 = pd.DataFrame(BA)
prices1 = prices1.set_index('Date')
prices1 = prices1['December-31-2017':'Jan-01-2013']
prices1.tail()
```

```python
#Putting Data Together
BAdata = prices1.join(clean_options1)
BAdata.head()
```

```python
#Putting Data into a Pickle for EDA
BAdata.to_pickle('BAdata.pickle')
```
